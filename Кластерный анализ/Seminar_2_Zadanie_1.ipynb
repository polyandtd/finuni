***Для указанных наборов данных провести реализацию метода главных компонент.***

```python
df
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
```

```python
df2 = df.drop('type', axis=1)
```

```python
df2.corr()
```

```python
df2_scaled = df2.copy()
df2_scaled= sc.fit_transform(df2_scaled)
df2_scaled = pd.DataFrame(df2_scaled, columns = df2.columns)
```

```python
df2_scaled
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
pca.fit(df2_scaled)
print(pca.explained_variance_ratio_)
```

```python
pca = PCA(n_components=1)
pca.fit(df2_scaled)
print(pca.explained_variance_ratio_)
```

***Факторный анализ***

```python
from sklearn.decomposition import FactorAnalysis

fa = FactorAnalysis(n_components=1)
fa.fit(df2_scaled)
```

```python
pd.DataFrame(fa.components_, columns=df2.columns)
```

```python
print('Uniqueness:')
pd.Series(fa.noise_variance_, df2.columns)
```
