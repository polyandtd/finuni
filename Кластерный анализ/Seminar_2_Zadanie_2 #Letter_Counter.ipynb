***2) Подсчитать, сколько раз встречается каждая из русских букв во всех текстовых файлах, лежащих в папке 'text'. При реализации выделить функцию, которая суммирует результаты обработки отдельных файлов. Определить за какое время решается задача для всех файлов из папки 'text'.***

```python
import time
```

```python
start_time = time.time()

books = ['Tolstoy Lev. Voyna i mir. Kniga 1 - BooksCafe.Net',
        'Tolstoy Lev. Anna Karenina - BooksCafe.Net',
       'Dostoevskiy Fedor. Besy - BooksCafe.Net',
         'Dostoevskiy Fedor. IDIOT - BooksCafe.Net',
         'Dostoevskiy Fedor. Igrok - BooksCafe.Net',
         'Dostoevskiy Fedor. Prestuplenie i nakazanie - BooksCafe.Net',
         'Dostoevskiy Fedor. Tom 7. Besy - BooksCafe.Net',
         'Tolstoy Lev. Voyna i mir. Kniga 2 - BooksCafe.Net',
         'Достоевский Федор. Братья Карамазовы - royallib.ru'
        ]

#переменная для хранения значений в каждой из книг
ini_dict=[]

for book in books:
    
    f = open('text/text/' + book + '.txt', encoding = 'cp1251')
    lines = f.read()
    f.close()
    lines = lines.lower()
    
    ini_dict.append(letter_counter(lines))
    print(book.split('-')[0], letter_counter(lines), sep='\n')
    print()


result = {} 
for d in ini_dict: 
    for k in d.keys(): 
        result[k] = result.get(k, 0) + d[k] 
  
  
print("Количество каждой русской буквы во всех книгах: ", result, sep='\n')
print()
print("Code run in --- %s seconds ---" % round(time.time() - start_time, 5))
```
