```python
df = pd.read_csv('sp500hst.txt', names=['date', 'name', 'open', 'high', 'low', 'close', 'volume'])
df.head()
```

```python
df.info()
```

```python
df.date = df.date.astype(str)
```

```python
def get_year(row):
        return row['date'][:4]
    
def get_month(row):
        return row['date'][4:6]
    
def get_day(row):
        return row['date'][6:]
```

```python
df['year'] = df.apply(get_year, axis=1)
df['month'] = df.apply(get_month, axis=1)
df['day'] = df.apply(get_day, axis=1)
```

```python
df.drop('date', axis=1, inplace=True)
```
