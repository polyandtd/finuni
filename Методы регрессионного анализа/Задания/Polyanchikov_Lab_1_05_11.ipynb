#### Выведите только те строки таблицы, даты создания магазина которых последние «пять» лет

```python
def to_ts(row):
    
    return time.mktime(row.date.timetuple())
```

```python
df['timestamp'] = df.apply(to_ts, axis=1)
```

```python
def get_year(row):
    
    return datetime.fromtimestamp(row.timestamp).year
```

```python
df['year'] = df.apply(get_year, axis=1)
```

```python
df.head()
```

```python
df.loc[df.date.dt.year > 2014, :]
```

#### Выведите только те строки значения таблицы, даты создания магазина которых последние «пять лет» и они находятся не в Москве

```python
df.loc[(df.date.dt.year > 2014) & (df.city != 'Moscow'), :]
```
