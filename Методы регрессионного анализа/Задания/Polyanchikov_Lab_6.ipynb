```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv('bank-additional-full.csv', sep=';')
df.head()
```

```python
df['y'] = df['y'].map({'yes': 1, 'no': 0})
```

```python
df.head()
```

```python
df.y.value_counts()
```

```python
df.info()
```

```python
df.shape
```

```python
df = pd.get_dummies(df)
```

```python
df.head()
```

```python
y = df['y']
X = df.drop('y', axis=1)
cols = X.columns
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X)
```

```python
X = pd.DataFrame(X, columns = cols)
```

```python
df2 = X.join(y)
```

```python
df2.corr()['y'].sort_values(ascending=False).reset_index()
```

```python
X.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, test_size = 0.5)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
```

```python
logistic = LogisticRegression(random_state=42)
svc = SVC(random_state=42)
```

```python
logistic.fit(X_train, y_train), svc.fit(X_train, y_train)
```

```python
logistic_pred = logistic.predict(X_valid)
svc_pred = svc.predict(X_valid)
```

```python
from sklearn.metrics import roc_auc_score
```

```python
roc_auc_score(y_valid, logistic_pred)
```

```python
roc_auc_score(y_valid, svc_pred)
```
