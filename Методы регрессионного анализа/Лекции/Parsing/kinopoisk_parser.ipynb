```python
import requests
from bs4 import BeautifulSoup
import csv
import pandas as pd
```

```python
def get_html(url):
    r = requests.get(url)
    return r.text
```

```python
def write_csv(data):
    with open('kinopoisk.csv', 'a') as f:
        writer = csv.writer(f)
        writer.writerow((data['rus_title'],
                         data['original_title'],
                         data['country'],
                         data['genre'], 
                         data['rating'],
                         data['votes'],
                         data['position']))
```

```python
def get_page_data(html):
    soup = BeautifulSoup(html, 'lxml')
    divs = soup.find('div', class_='selection-list selections-seo-page__list')
    films = divs.find_all('div', class_='selections-film-item selections-film-item_is-watched_null selection-list__film')
    for film in films:
        try:
            div = film.find('p', class_ = 'selection-film-item-meta__name')
            rus_title = div.text.strip()
        except:
            rus_title = ''
        try:
            div = film.find('p', class_ = 'selection-film-item-meta__original-name')
            original_title = div.text.strip()
        except:
            original_title = ''
        try:
            div = film.findAll('span', class_="selection-film-item-meta__meta-additional-item")[0]
            country = div.text.strip()
        except:
            country = ''
        try:
            div = film.findAll('span', class_="selection-film-item-meta__meta-additional-item")[1]
            genre = div.text.strip()
        except:
            genre = ''
        try:
            div = film.find('span', class_="rating__value rating__value_positive")
            rating = div.text.strip()
        except:
            rating = ''
        try:
            div = film.find('span', class_="rating__count")
            votes = div.text.strip()
        except:
            votes = ''
        try:
            div = film.find('span', class_="selection-film-item-position")
            position = div.text.strip()
        except:
            position = ''
        data = {
            'rus_title':rus_title,
            'original_title':original_title,
            'country':country,
            'genre':genre, 
            'rating':rating,
            'votes':votes,
            'position':position
       }
        write_csv(data)
```

```python
def main():
    base_url = "https://www.kinopoisk.ru/lists/navigator/"   
    page_part = "?page="
    query_par = "country-1/"
    for i in range(1,5):
        
        url_gen = base_url + query_par + page_part + str(i)
        print(url_gen)
        html = get_html(url_gen)
        get_page_data(html)
```

```python
if __name__ == '__main__':
    main()
```
